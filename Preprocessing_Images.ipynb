```python
# Dataset from http://www.cs.columbia.edu/CAVE/databases/pubfig/download/
import os
import urllib.request
import json
from os import listdir
import matplotlib.pyplot as plt
import cv2

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
url_file= open('dev_urls.txt' ,'r')
file_info=[]
line_idx = 1
for idx,line in enumerate(url_file):
    if idx > 1:
        words = str.split(line,'\t')
        filename = words[0].replace(' ','')+words[1]
        url = words[2]
        facedim = words[3]
        """
        if not os.path.exists('dataset/'+words[0]): os.makedirs('dataset/'+words[0])
        try:
            file = urllib.request.urlretrieve(url,'dataset/'+words[0]+'/'+filename+'.jpg')
        except:
            continue
        """
        file_info.append([filename,url,facedim])
        
        

url = 'http://1.bp.blogspot.com/_Y7rzCyUABeI/SNIltEyEnjI/AAAAAAAABOg/E1keU_52aFc/s400/ash_abhishek_365x470.jpg'
urllib.request.urlretrieve(url,'blah.jpg')
```

```python
file_list = [item[0] for item in file_info]
faces_idx = 1;
for folder in listdir('dataset/'):
    for image in listdir('dataset/'+folder):
            for idx,filename in enumerate(file_list):
                if filename == image.split('.')[0]:
                    size  = file_info[idx][2].split(',')
                    print(size,filename,image,folder)
                    try:
                        img = cv2.imread('dataset/'+folder+'/'+image)
                        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                    except cv2.error as e:
                        continue
                    faces = face_cascade.detectMultiScale(gray, 1.3, 2)
                    
                    if faces != []:
                        for (x,y,w,h) in faces:
                            print(x,y,w,h)
                            cropped_img =  img[y:y+h, x:x+w]
                            if not os.path.exists('dataset_cropped/'+folder): os.makedirs('dataset_cropped/'+folder)
                            cv2.imwrite(("dataset_cropped/{}/{}{}.jpg".format(folder,filename,faces_idx)),cropped_img)
                        faces_idx = 1;

                        
            
#eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```
